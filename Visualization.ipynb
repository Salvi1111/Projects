```python
df_country= pd.read_excel("Country-Code.xlsx")
df_country.head()
```

Merge df and df_country : on helps in getting which column we want to merge and how- which table to merge in left or right

```python
final_df= pd.merge(df, df_country, on= "Country Code", how= "left")  #left= df
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_name= final_df.Country.value_counts().index
country_name
```

```python
country_value=final_df.Country.value_counts().values
```

```python
# Creating Pie chart for top 3 countries

plt.pie(country_value[:3], labels= country_name[:3])
plt.show()
```

```python
#get percentage values
plt.pie(country_value[:3], labels= country_name[:3], autopct="%1.2f%%")
plt.show()
```

Observation: Maximum transactions of Zomato are from India. After that USA and then UK.

```python
# to understand some columns we'll use grouby

final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size()
```

```python
rating=final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
rating
```

''' Observation: 
1. Whenever the rating is b/w 4.5 to 4.9 it indicated that its excellent
2. When ratings are b/w From 4.0 to 4.5 it indicated that its V good
3. When ratings are b/w From 3.5 to 3.9 it indicated that its good
4. When ratings are b/w From 3.0 to 3.4 it indicated that its Avg
5. When ratings are b/w From 2.5 to 2.9 it indicated that its Avg '''
